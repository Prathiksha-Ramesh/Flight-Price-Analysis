```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head()
```

```python
#get the basic info about the data:

df.info()
```

```python
df.describe()
```

```python
## feature engineering:

df['Date']=df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
#drop the date of journey:

df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df["Arrival_hour"]=df['Arrival_Time'].str.split(':').str[0]
df["Arrival_minute"]=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_minute']=df['Arrival_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df["Dep_hour"]=df['Dep_Time'].str.split(':').str[0]
df["Dep_minute"]=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_minute']=df['Dep_minute'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
# Extract hours
df['Duration_hr'] = df['Duration'].str.split(' ').str[0].str.replace('h', '')

# Extract minutes
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.replace('m', '')

# Handle cases where there are no minutes or hours (e.g., "5h" or "30m")
df['Duration_hr'] = df['Duration_hr'].fillna('0')
df['Duration_min'] = df['Duration_min'].fillna('0')
```

```python
df['Duration_hr']=df['Duration_hr'].astype(int)
df['Duration_min']=df['Duration_min'].astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encoded_df=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df=pd.concat([df,encoded_df],axis=1)
```
